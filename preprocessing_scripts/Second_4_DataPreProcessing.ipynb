```python
import pandas as pd
import requests
try:
    from StringIO import StringIO
except ImportError:
    from io import StringIO
```

# Loading 

```python
def load_DF(url, skip):
  data = StringIO(requests.get(url).text)
  return pd.read_csv(data, skiprows=range(0,skip))
```

```python
oil_rents_df = load_DF("https://raw.githubusercontent.com/mohannashahrad/Borealis_AI_Plant_Tree_Project/main/Data/Oil%20rents%20(%25%20of%20GDP)/API_NY.GDP.PETR.RT.ZS_DS2_en_csv_v2_3054193.csv", 4)

agriculture_outlook_df = load_DF("https://raw.githubusercontent.com/mohannashahrad/Borealis_AI_Plant_Tree_Project/main/Data/Agricultural%20Outlook%202021-2030.csv", -1)
```
