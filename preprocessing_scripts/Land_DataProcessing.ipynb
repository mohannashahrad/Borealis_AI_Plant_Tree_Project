```python
# Import all the required libraries
import pandas as pd
import requests
try:
    from StringIO import StringIO
except ImportError:
    from io import StringIO
```

```python
# Define all the functions

def load_DF(url, skip):
  data = StringIO(requests.get(url).text)
  return pd.read_csv(data, skiprows = skip)
```

#Load datasets

```python
total_df = load_DF('https://raw.githubusercontent.com/mohannashahrad/Borealis_AI_Plant_Tree_Project/main/Data/Land_Total_Data_1960_2020.csv', range(0, 3))
total_df.head()
```

```python
forest_df = load_DF('https://raw.githubusercontent.com/mohannashahrad/Borealis_AI_Plant_Tree_Project/main/Data/Land_Forest_Data_1960_2020.csv', range(0, 3))
forest_df.head()
```

```python
agrc_df = load_DF('https://raw.githubusercontent.com/mohannashahrad/Borealis_AI_Plant_Tree_Project/main/Data/Land_Agriculture_Data_1960_2020.csv', range(0, 3))
agrc_df.head()
```

```python
total_df.drop(['Country Code', 'Indicator Name', 'Indicator Code', 'Unnamed: 65'], axis=1, inplace=True)
total_df = pd.melt(total_df, id_vars='Country Name', var_name='Time', value_name='Total Land (m2)')
total_df.rename(columns={"Country Name": "Country"}, inplace=True)
total_df['Time'] = total_df['Time'].astype(int)
total_df = total_df[total_df.Time >= 1990]
total_df = total_df[total_df.Time <= 2020]
display(total_df)
```

```python
forest_df.drop(['Country Code', 'Indicator Name', 'Indicator Code', 'Unnamed: 65'], axis=1, inplace=True)
forest_df = pd.melt(forest_df, id_vars='Country Name', var_name='Time', value_name='Forest (%)')
forest_df.rename(columns={"Country Name": "Country"}, inplace=True)
forest_df['Time'] = forest_df['Time'].astype(int)
forest_df = forest_df[forest_df.Time >= 1990]
forest_df = forest_df[forest_df.Time <= 2020]
display(forest_df)
```

```python
agrc_df.drop(['Country Code', 'Indicator Name', 'Indicator Code', 'Unnamed: 65'], axis=1, inplace=True)
agrc_df = pd.melt(agrc_df, id_vars='Country Name', var_name='Time', value_name='Agriculture (%)')
agrc_df.rename(columns={"Country Name": "Country"}, inplace=True)
agrc_df['Time'] = agrc_df['Time'].astype(int)
agrc_df = agrc_df[agrc_df.Time >= 1990]
agrc_df = agrc_df[agrc_df.Time <= 2020]
display(agrc_df)
```

```python
total_index = total_df.set_index(['Country', 'Time'])
forest_index = forest_df.set_index(['Country', 'Time'])
agrc_index = agrc_df.set_index(['Country', 'Time'])
merged_df = total_index.join([forest_index, agrc_index], how='outer')
merged_df.head()
```

```python
merged_df.reset_index(inplace=True)
merged_df.to_csv("land.csv")
```
