```python
# Import all the required libraries
import pandas as pd
import requests
try:
    from StringIO import StringIO
except ImportError:
    from io import StringIO
```

```python
# Define all the functions

def load_DF(url, skip):
  data = StringIO(requests.get(url).text)
  return pd.read_csv(data, skiprows=range(1,skip))
```

```python
# Loading Datasets

# Load GDP Dataset
gdp_df = load_DF('https://raw.githubusercontent.com/mohannashahrad/Borealis_AI_Plant_Tree_Project/main/Data/Real%20GDP%20long-term%20forecast.csv', -1)
display(gdp_df)

# Load GDP Dataset
population_df = load_DF('https://raw.githubusercontent.com/mohannashahrad/Borealis_AI_Plant_Tree_Project/main/Data/migration_population.csv', 2774)
display(population_df)
```

```python
gdp_df.drop(['INDICATOR', 'SUBJECT', 'MEASURE', 'FREQUENCY', 'Flag Codes'], axis=1, inplace=True)
gdp_df.rename(columns={"LOCATION": "Country Code", "TIME": "Year", "Value": "GDP (M_USD)"}, inplace=True)
gdp_df.head()
```

```python
population_df.drop(['net_migration', 'migration_perc', 'iso2c', 'adminregion', 'incomeLevel', 'lendingType', 'capitalCity', 'longitude', 'latitude'], axis=1, inplace=True)
population_df.rename(columns={"country": "Country Name", "iso3c": "Country Code", "year": "Year", "population": "Population", "region": "Region", "pop_density": "Pop Density"}, inplace=True)
population_df.head()
```

```python
gdp_index = gdp_df.set_index(['Country Code', 'Year'])
pop_index = population_df.set_index(['Country Code', 'Year'])
merged_df = gdp_index.join(pop_index, how='outer')
merged_df.head()
```

```python
merged_df.reset_index(inplace=True)
merged_df.to_csv("pop_gdp.csv")
```
